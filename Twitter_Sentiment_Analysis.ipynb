```python
# import all relevant libraries 

import json 
import twitter
import pandas as pd
import nltk
import re
```

```python
# Keys and token from Twitter Dev
consumer_key = ''
consumer_secret = ''
oauth_token = ''
oauth_token_secret = ''

# authentication
auth = twitter.oauth.OAuth(oauth_token,oauth_token_secret,
    consumer_key,consumer_secret)

twitter_api = twitter.Twitter(auth=auth)

def twitter_search(twitter_api, q,lang, max_results=500, **kw):
    search_results = twitter_api.search.tweets(q=q, count=100, **kw)
    statuses = search_results['statuses']
    max_results = min(1000, max_results)
    for _ in range(10): # 10*100 = 1000 
        try:
            next_results = search_results['search_metadata']['next_results'] 
        except KeyError as e: # No more results when next_results doesn't exist
            break
        kwargs = dict([ kv.split('=') for kv in next_results[1:].split("&") ]) 
        search_results = twitter_api.search.tweets(**kwargs)
        statuses += search_results['statuses'] 
        if len(statuses) > max_results:
            break 

    return statuses

q = "Iphone X"
lang = "en"

results = twitter_search(twitter_api, q, lang, max_results=10)
print(results)
```

```python
tweets_df = pd.DataFrame(results)
#tweets_df
tweets_df["lang"]
tweets_df = tweets_df.loc[tweets_df['lang'] == "en"]
```

```python
tweet_text = []
def clean_tweet(tweet):
    for i in tweet:
        '''
        Utility function to clean tweet text by removing links, special characters
        using simple regex statements.
        '''
        tweet_text.append( ' '.join(re.sub("(@[A-Za-z0-9]+)|([^0-9A-Za-z \t])|(\w+:\/\/\S+)", " ", i).split()))
clean_tweet(tweets_df["text"])
```

```python
# function to remove "RT" from tweet
clean_tweets =[]
def remove_retweets():  
    for i in tweet_text:
         rt = 'RT'
         Tweet = i.replace(rt,'')
         clean_tweets.append(Tweet)
remove_retweets()        
```

```python
def format_sentence(sent):
    return({word: True for word in nltk.word_tokenize(sent)})
```

```python
pos = []
with open("./pos_tweets.txt") as f:
    for i in f: 
        pos.append([format_sentence(i), 'pos'])
pos
```

```python
neg = []
with open("./neg_tweets.txt") as f:
    for i in f: 
        neg.append([format_sentence(i), 'neg'])
        
neg
```

```python
training = pos[:int((.9)*len(pos))] + neg[:int((.9)*len(neg))]
```

```python
test = pos[int((.1)*len(pos)):] + neg[int((.1)*len(neg)):]
test
```

```python
# import Naive Bayes Classifer
from nltk.classify import NaiveBayesClassifier

classifier = NaiveBayesClassifier.train(training)
```
